```python
#task-2
class Student():
  def __init__(self,n,id,g):
    self.n=n
    self.id=id
    self.g=g
  def setId(self,id):
    self.id=id
class Department():
  def __init__(self,d):
    self.d=d
    self.st=[]
    self.stid=[]
  def findStudent(self,idx):
    for i in self.st:
      if idx == i.id:
        print(f'Student info:')
        print(f'Student Name: {i.n}')
        print(f'ID: {i.id}')
        print(f'CGPA: {i.g}')
        break
      else:
        print(f"Student with this ID doesn't exist, Please give a valid ID")
  def addStudent(self,*ad):
    for i in ad:
      if i.id in self.stid:
        print(f'Student with the same ID already exists, Please try with another ID')
      else:
        print(f'Welcome to CSE department, {i.n}')
        self.st.append(i)
        self.stid.append(i.id)
  def details(self):
    print(f'Department Name: {self.d}')
    print(f'Number of student:{len(self.st)}')
    print(f'Details of the students:')
    for i in self.st:
      print(f'Student name: {i.n}, ID: {i.id}, cgpa: {i.g}')

s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```

```python
#task-3
class Vaccine():
  def __init__(self,t,c,d):
    self.t=t
    self.c=c
    self.d=d

class Person():
  def __init__(self,n,a,p='General Citizen'):
    self.n=n
    self.a=a
    self.p=p
    self.ty=None
    self.td='1st Dose'
  def pushVaccine(self,ty,td='1st Dose'):
    if (self.ty!=ty) and (self.ty!=None):
      print(f'Sorry {self.n}, you canâ€™t take 2 different vaccines')
    elif self.ty==None:
      print(f'{td} done for {self.n}')
    elif self.td=='2nd Dose':
      print(f'2nd dose done for {self.n}')
    self.ty=ty
    self.td=td
  def showDetail(self):
    for i in self.ty:
      print(f'Name: {self.n} Age: {self.a} Type: {self.p}')
      print(f'Vaccine name: {i.t}')
      if self.td=='1st Dose':
        print(f'1st dose: Given')
        print(f'2nd dose: Please come after 60 days')
      else:
        print(f'1st dose: Given')
        print(f'2nd dose: Given')


astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
